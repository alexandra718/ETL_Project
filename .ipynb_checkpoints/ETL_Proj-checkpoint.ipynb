```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine
```

```python
debt_file = "Resources/Overall_Lowest_Debt.csv"
debt_df = pd.read_csv(debt_file)
debt_df.head()
```

```python
#getting list of all univ names
debt_df['Institution Name'].unique()
```

```python
salary_file = "Resources/salaries-by-college-type.csv"
salary_df = pd.read_csv(salary_file)
salary_df.head(12)
```

```python
salary_df['School Name'].unique()
```

```python
#drop NaN for debt df add.copy() to save this new df as seperate 
debt_df2=debt_df.dropna().copy()

debt_df2.head()
```

```python
#getting the necessary information from the dataframe debt
debt_df3=debt_df2[['Institution Name','Average Debt Per Borrower']].copy()

debt_df3.head()
```

```python
#salary_df remove unncessasery columns
salary_df2=salary_df[['School Name', 'Starting Median Salary', 'Mid-Career Median Salary']].copy()
salary_df2.dropna()

salary_df2.head(14)
```

```python
#change name to what works
salary_df2['School Name'].replace("\\((.*?)\\)", "", inplace=True, regex=True)

salary_df2.head()
```

```python
#debt transformed

debt_transformed=debt_df3.copy()

debt_transformed=debt_transformed.rename(columns={"Institution Name": "university_name",
                                                          "Average Debt Per Borrower": "avg_debt_per_borrower"})

debt_transformed.reset_index()

debt_transformed.head()
```

```python
#salary transformed

salary_transformed=salary_df2.copy()

salary_transformed=salary_transformed.rename(columns={'School Name': "university_name",
                                                     "Starting Median Salary":"starting_med_salary",
                                                     "Mid-Career Median Salary":"mid_career_med_salary"})

salary_transformed.head()
```
